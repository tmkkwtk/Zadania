# Zadanie: statystyki

```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
# parametry rozkładu normalnego
mean = 3
std_dev = 1
```

```python
# generowanie danych
dist_100 = np.random.normal(mean, std_dev, 100)
dist_10000 = np.random.normal(mean, std_dev, 10000)
```

```python
# statystyki opisowe dla dist_100
mean_dist_100 = np.round(np.mean(dist_100), 2)  # średnia
median_dist_100 = np.round(np.median(dist_100), 2)  # mediana
mode_dist_100 = np.round(stats.mode(dist_100)[0], 2)  # moda
q0_100 = np.round(np.quantile(dist_100, 0.0), 2)  # wartość minimalna
q1_100 = np.round(np.quantile(dist_100, 0.25), 2)  # 1 kwartyl
q3_100 = np.round(np.quantile(dist_100, 0.75), 2)  # 3 kwartyl
q4_100 = np.round(np.quantile(dist_100, 1.0), 2)  # 4 kwartyl
range_dist_100 = np.max(dist_100) - np.min(dist_100)  # rozstęp
iqr_100 = q3_100 - q1_100  # rozstęp międzykwartylowy
variance_dist_100 = np.round(np.var(dist_100, ddof=1), 2)  # wariancja
standard_deviation_dist_100 = np.round(np.std(dist_100, ddof=1), 2)  # odchylenie standardowe

# statystyki opisowe dla dist_10000
mean_dist_10000 = np.round(np.mean(dist_10000), 2)  # średnia
median_dist_10000 = np.round(np.median(dist_10000), 2)  # mediana
mode_dist_10000 = np.round(stats.mode(dist_10000)[0], 2)  # moda
q0_10000 = np.round(np.quantile(dist_10000, 0.0), 2)  # wartość minimalna
q1_10000 = np.round(np.quantile(dist_10000, 0.25), 2)  # 1 kwartyl
q3_10000 = np.round(np.quantile(dist_10000, 0.75), 2)  # 3 kwartyl
q4_10000 = np.round(np.quantile(dist_10000, 1.0), 2)  # 4 kwartyl
range_dist_10000 = np.max(dist_10000) - np.min(dist_10000)  # rozstęp
iqr_10000 = q3_10000 - q1_10000  # rozstęp międzykwartylowy
variance_dist_10000 = np.round(np.var(dist_10000, ddof=1), 2)  # wariancja
standard_deviation_dist_10000 = np.round(np.std(dist_10000, ddof=1), 2)  # odchylenie standardowe
```

```python
# wyniki dla każdego zbioru
print("Statystyki dla 100 próbek:")
print(f"Średnia: {mean_dist_100}, Mediana: {median_dist_100}, Moda: {mode_dist_100}")
print(f"Min: {q0_100}, 1Q: {q1_100}, 3Q: {q3_100}, Max: {q4_100}, Rozstęp: {range_dist_100}")
print(f"IQR: {iqr_100}, Wariancja: {variance_dist_100}, Odchylenie standardowe: {standard_deviation_dist_100}")
print()

print("Statystyki dla 10000 próbek:")
print(f"Średnia: {mean_dist_10000}, Mediana: {median_dist_10000}, Moda: {mode_dist_10000}")
print(f"Min: {q0_10000}, 1Q: {q1_10000}, 3Q: {q3_10000}, Max: {q4_10000}")
print(f"Rozstęp: {range_dist_10000}, IQR: {iqr_10000}, Wariancja: {variance_dist_10000}")
print(f"Odchylenie standardowe: {standard_deviation_dist_10000}")

# porównanie
print("\nPorównanie wyników dla 100 próbek i 10,000 próbek:")

# średnia
if mean_dist_100 == mean_dist_10000:
    print(f"Średnia dla obu zbiorów jest taka sama: {mean_dist_100}.")
else:
    print(f"Średnia dla 100 próbek: {mean_dist_100}, dla 10,000 próbek: {mean_dist_10000}.")

# mediana
if median_dist_100 == median_dist_10000:
    print(f"Mediana dla obu zbiorów jest taka sama: {median_dist_100}.")
else:
    print(f"Mediana dla 100 próbek: {median_dist_100}, dla 10,000 próbek: {median_dist_10000}.")

# odchylenie standardowe
if standard_deviation_dist_100 == standard_deviation_dist_10000:
    print(f"Odchylenie standardowe dla obu zbiorów jest takie same: {standard_deviation_dist_100}.")
else:
    print(f"Odchylenie standardowe dla 100 próbek: {standard_deviation_dist_100}, dla 10,000 próbek: {standard_deviation_dist_10000}.")

# wariancja
if variance_dist_100 == variance_dist_10000:
    print(f"Wariancja dla obu zbiorów jest taka sama: {variance_dist_100}.")
else:
    print(f"Wariancja dla 100 próbek: {variance_dist_100}, dla 10,000 próbek: {variance_dist_10000}.")

# rozstęp
if range_dist_100 == range_dist_10000:
    print(f"Rozstęp dla obu zbiorów jest taki sam: {range_dist_100}.")
else:
    print(f"Rozstęp dla 100 próbek: {range_dist_100}, dla 10,000 próbek: {range_dist_10000}.")

# wykres
plt.plot(np.arange(len(dist_100)), np.sort(dist_100), label="100 próbek")
plt.plot(np.arange(len(dist_10000)), np.sort(dist_10000), label="10,000 próbek")
plt.xlabel('Index')
plt.ylabel('Wartość')
plt.title('Rozkład normalny (średnia 3, odchylenie 1)')
plt.legend()
plt.show()
```
