```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, root_mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, LSTM, GRU

# Ustawienia
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
os.environ["OMP_NUM_THREADS"] = "1"

# Wczytanie danych
dataset = pd.read_csv('airline-passengers.csv')
dataset['Month'] = pd.to_datetime(dataset['Month'])
dataset.set_index(['Month'], inplace=True)

# Wizualizacja danych
plt.figure(figsize=(16, 9))
plt.plot(dataset['Passengers'], label='Passengers')
plt.title('Passengers over Time')
plt.legend()
plt.show()

# Skalowanie danych
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(dataset)

# Przygotowanie zbiorów treningowego, walidacyjnego i testowego
train_size = int(len(data_scaled) * 0.7)
val_size = int(len(data_scaled) * 0.15)
test_size = len(data_scaled) - train_size - val_size

train, val, test = (
    data_scaled[:train_size],
    data_scaled[train_size:train_size + val_size],
    data_scaled[train_size + val_size:]
)

def create_dataset(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

# Grid Search dla hiperparametrów
best_rmse = float('inf')
best_model = None
best_params = None

for look_back in range(1, 13):
    # Tworzenie dynamicznych zbiorów danych dla bieżącego look_back
    X_train, y_train = create_dataset(train, look_back)
    X_val, y_val = create_dataset(val, look_back)
    X_test, y_test = create_dataset(test, look_back)

    # Reshape danych do formatu (samples, timesteps, features)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_val = np.reshape(X_val, (X_val.shape[0], X_val.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    for units in range(1, 13):
        print(f'Testing look_back={look_back}, units={units}')
        
        # Budowa modelu
        model = Sequential()
        model.add(LSTM(units, input_shape=(look_back, 1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mean_squared_error')
        
        # Trening modelu
        model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val))
        
        # Ewaluacja na zbiorze walidacyjnym
        val_pred = model.predict(X_val)
        val_pred_rescaled = scaler.inverse_transform(val_pred)
        y_val_rescaled = scaler.inverse_transform(y_val.reshape(-1, 1))
        rmse = mean_squared_error(y_val_rescaled, val_pred_rescaled, squared=False)
        
        print(f'RMSE: {rmse}')
        
        # Zapis najlepszego modelu
        if rmse < best_rmse:
            best_rmse = rmse
            best_model = model
            best_params = (look_back, units)

print(f'Best Model: look_back={best_params[0]}, units={best_params[1]}, RMSE={best_rmse}')

# Wizualizacja prognoz i wartości rzeczywistych
def plot_predictions(model, X, y, dataset, title):
    predictions = model.predict(X)
    predictions_rescaled = scaler.inverse_transform(predictions)
    y_rescaled = scaler.inverse_transform(y.reshape(-1, 1))
    
    plt.figure(figsize=(16, 9))
    plt.plot(dataset.index[-len(y_rescaled):], y_rescaled, label='True Values', color='blue')
    plt.plot(dataset.index[-len(predictions_rescaled):], predictions_rescaled, label='Predictions', color='red')
    plt.title(title)
    plt.legend()
    plt.show()

# Wizualizacja na zbiorze testowym
plot_predictions(best_model, X_test, y_test, dataset[-len(y_test):], 'Test Set Predictions')
```
